# Propagação de ondas planas num espaço ilimitado (domínio do tempo)

## Objetivo

Visualizar ondas planas propagando-se para baixo no meio homogêneo da terra com excitação por impulso. Com os dois aplicativos: a) Plane wave app (aplicativo de onda plana) e b) Profile app (aplicativo de perfil), entender os conceitos fundamentais da propagação de ondas planas no domínio do tempo.


## Set up

As soluções das ondas planas EM para terra homogênea com excitação impulsiva podem ser expressas como

$$\mathbf{e}(t) = -E_0 \frac{(\mu\sigma)^{1/2}z}{2 \pi^{1/2} t^{3/2}} e^{-\mu\sigma z^2 / (4t)} \mathbf{u_x}$$

$$ \mathbf{h}(t) = E_0 \sqrt{\dfrac{\sigma}{\pi\mu t}}\, e^{-\mu\sigma z^2/4t} \, \mathbf{u_y} $$

Observe que esse desvio é baseado em uma aproximação quase estática, que ignora as correntes de deslocamento. Para derivação detalhada, consulte as notas de aulas.
